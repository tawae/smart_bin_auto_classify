```python
import sqlite3
import os
from google.colab import drive
import pandas as pd
```

```python

drive.mount('/content/drive/', force_remount=True)

DRIVE_PATH = '/content/drive/MyDrive/SmartTrash'
DB_PATH = os.path.join(DRIVE_PATH, 'waste_log.db')

# Kết nối CSDL
conn = sqlite3.connect(DB_PATH)


table_to_check = 'cardboard'

try:
    # Đọc 10 dòng mới nhất từ bảng 'paper'
    query = f"SELECT * FROM {table_to_check} ORDER BY timestamp DESC LIMIT 10"

    # Sử dụng Pandas để đọc và in ra bảng
    df = pd.read_sql_query(query, conn)

    print(f"--- 10 bản ghi mới nhất trong bảng '{table_to_check}' ---")
    tab = pd.DataFrame(df)
    display(tab)

except Exception as e:
    print(f"Lỗi khi truy vấn bảng '{table_to_check}': {e}")

conn.close()
```
